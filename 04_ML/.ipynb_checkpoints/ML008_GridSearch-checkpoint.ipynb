```python
# ML8_GridSearch
import scipy.stats
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV # GridSearch
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
# データロード
dt = load_digits()
# データ分割
ax, ex, ay, ey = train_test_split(dt.data, dt.target, random_state=0)
# パラメタ辞書作成
mp = {SVC(): {
    "kernel": ["linear", "poly", "rbf", "sigmoid"],
    "C": [10 ** i for i in range(-5, 5)], # GridSearch
    "decision_function_shape": ["ovr", "ovo"],
    "random_state": [0] # GridSearch
}}
# 定数作成
max_score = 0
best_param = None 
# パラメタ辞書から値取り出し
for m, p in mp.items():
    # グリッドサーチ
    m = GridSearchCV(m, p)
    # 適合
    m.fit(ax, ay)
    # 予測
    py = m.predict(ex)
    # スコア算出
    sc = f1_score(ey, py, average="micro")
    # スコアがベストを更新したら
    if max_score < sc:
        # ベストスコア更新
        max_score = sc
        # モデル名保存
        best_model = m.__class__.__name__
        # ベストパラメタ保存
        best_param = m.best_params_
        
# 結果1（適用あり）（ベストパラメタ・ベストスコア）
print("適用あり")
print(best_param)
print(max_score)

# 結果2（適用なし）（テストスコア）
m0 = SVC()
m0.fit(ax, ay)
print("適用なし")
print(m0.score(ex,ey))
```
