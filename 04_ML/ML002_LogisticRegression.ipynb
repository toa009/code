```python
# ML2_LogisticRegression

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, random_state=0)

ax, ex, ay, ey = train_test_split(X, y, random_state=0)

# LogisticRegression ----------------------------------------

from sklearn.linear_model import LogisticRegression
m = LogisticRegression()

# -----------------------------------------------------------------

m.fit(ax, ay)
py = m.predict(ex)

plt.scatter(X[:,0], X[:,1], c=y, marker='.')

Xi = np.linspace(-10, 10)
Y = -( Xi * m.coef_[0][0] + m.intercept_) / m.coef_[0][1]
plt.plot(Xi, Y)

plt.xlim(min(X[:, 0]) - 0.5, max(X[:, 0]) + 0.5)
plt.ylim(min(X[:, 1]) - 0.5, max(X[:, 1]) + 0.5)
plt.axes().set_aspect("equal", "datalim")
```
