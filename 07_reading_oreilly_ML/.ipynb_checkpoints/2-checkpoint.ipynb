2.3.2.2 KNeighborsClassifierの解析
- 2次元平面を分類結果で色分けしてみる

```python
# k-最近傍法 x forgeデータセット x 分類境界を色分け
f,a1= plt.subplots(1,3, figsize=(10,3))
l = [1,3,9] # 最近傍点数は1,3,9の3種類をプロット
for n,a in zip(l, a1):
    m = KNeighborsClassifier(n_neighbors=n).fit(x,y)
    
    # 境界色塗り
    mglearn.plots.plot_2d_separator(m, x, fill=True, eps=0.5, ax=a, alpha=0.4)
    
    # 散布図
    mglearn.discrete_scatter(x[:,0], x[:,1], y, ax=a)
    a.set_title(n)
    
a1[0].legend();
```

```python
# -> 最近傍点数が増えると（1 -> 9)、境界線がなめらかになる（モデルの単純化）
# 　 （仮に最近傍点数＝全データ数とすると、全予測結果は同じで境界なし）
```

```python
# モデルの複雑さと汎化性能の関係をプロット
from sklearn.datasets import load_breast_cancer
d = load_breast_cancer()
ax,ex,ay,ey = train_test_split(d.data, d.target, stratify=d.target, random_state=66)

aa = [] # 訓練精度
ea = [] # テスト精度

ns = range(1, 11) # 最近傍点数を1から10まで試す

for n in ns:
    m = KNeighborsClassifier(n_neighbors=n).fit(ax,ay)
    aa.append(m.score(ax, ay))
    ea.append(m.score(ex, ey))

plt.plot(ns, aa, label="train acc")
plt.plot(ns, ea, label="test acc")
plt.legend();
```

```python
# -> 多数決を採る近傍点数を増やすと、訓練データに対する精度は落ちるが、
#    テストデータに対する精度は上がる（＝汎化性能は上がる）
#    ベストなのは n = 6 程度だが、最低でも0.88程度はあるため、場合によってはこれで充分
```

2.3.2.5 利点と欠点とパラメータ

k最近傍分類器の重要なパラメータ：
 1. 近傍点数（3や5が多い。調整必須）
 2. データポイント間の距離測度（デフォルトのユークリッド距離がほとんど）
 3. （前処理も重要）
 
長所：
 1. モデルが理解しやすい
 2. あまり調整しなくても精度充分な場合が多い
 3. データ小なら高速
 
短所：
 1. データ大で遅い（特徴量数でもサンプル数でも）
 2. 特徴量数が数百以上では機能しない
 3. 疎なデータセットでは特に性能悪い
 4. （→　これらの欠点を持たない線形モデルなどがあるため、実際には殆ど使われていない）
 



2.3.3 線形モデル
- 最初に試すべき。
- 非常に大きいデータセットに適する。
- 非常に高次元のデータに適する。
- 実用的に広く使われている
- 入力特徴量の線形関数を用いて予測

2.3.3.1 線形モデルによる回帰

予測式：　y = wx + b （ただし wx = w[0] x[0] + w[1] x[1] + ... )
 - x：　特徴量
 - w：　学習されたパラメータ（傾き。重み）
 - b：　学習されたパラメータ（切片）
 - y：　予測

```python
# 線形モデル x waveデータセット
mglearn.plots.plot_linear_regression_wave()
```

```python
# -> 傾き(w)≒0.4、切片≒0と図からも読み取れる。
```

線形回帰：
- 特徴量数 = 1：　直線
- 特徴量数 = 2：　平面
- 特徴量数 = 3以上：　超平面

線形モデルの長所：
- 特徴量が多数の場合に強力
（特に、訓練データ数　＜　特徴量数　
　：　どんなyでも完全に訓練データの線形関数としてモデル化可能（線形代数より））

線形回帰のアルゴリズム：
- 色々ある（w,bを学習する方法と、モデルの複雑さを制御する方法が違う）



2.3.3.2 線形回帰（通常最小二乗法）
- 平均二乗誤差：　差を二乗したものの平均値
- 通常最小二乗法(Ordinary Least Squares: OLS)：　最も単純 & 古典的な線形回帰手法
- 線形回帰：　
 - 訓練データにおいて、予測と真のターゲットyとの平均二乗誤差を最小にするw,bを求める。
 - パラメータなし（＝モデルの複雑さを制御する方法がない）


```python
# 線形回帰　x 単純なデータセット（waveデータセット（特徴量1つだけ））
from sklearn.linear_model import LinearRegression
x,y = mglearn.datasets.make_wave(n_samples=60)
ax,ex,ay,ey = train_test_split(x,y,random_state=42)
mL = LinearRegression().fit(ax,ay)

# 傾きと切片
print(mL.coef_)
print(mL.intercept_)
```

```python
# coef_     ：　傾き。配列。入力特徴量の数だけ値がある。（waveデータセットは特徴量1つなのでcoef_の要素は1つ）
# intercept_：　切片。浮動小数点値。常に単独。

# 末尾のアンダースコア：　訓練データから得られた属性であることを表す。
#                         （ユーザが設定したパラメータと区別するためのsklearnの仕様）
```

```python
# 訓練スコアとテストスコア
print(mL.score(ax, ay))
print(mL.score(ex, ey))
```

```python
# -> スコアは低め。（R^2：0.6は低め）
#    訓練スコアとテストスコアが近い　＝　未学習より。
```

線形モデル：
 - 特徴量数が少ない場合（低次元。単純なデータセット）：　未学習よりになる
 - 特徴量数が多い場合　（高次元。複雑なデータセット）：　過学習よりになる（回避方法が重要）

```python
# 線形回帰　x 複雑なデータセット（拡張bostonデータセット（特徴量数104））
x,y = mglearn.datasets.load_extended_boston()
ax,ex,ay,ey = train_test_split(x,y,random_state=0)
mL = LinearRegression().fit(ax,ay)
# 訓練スコアとテストスコア
print(mL.score(ax, ay))
print(mL.score(ex, ey))
```

```python
# -> 訓練スコアが良すぎ、テストスコア悪い　→　過学習（複雑なデータセットのため）

#　標準の線形回帰では複雑度を調整できない　-> リッジ回帰を試す。
```

2.3.3.3 リッジ回帰
 - 線形回帰の一種
 - 通常最小二乗法のwを、訓練データに対する予測だけでなく、他の制約に対しても最適化する
 - 制約：　wを0に近づけたい（特徴量が出力に与える影響を小さくしたい）（傾きを小さくしたい）（L2正則化）
 - 正則化：　モデルの単純化。過学習防止のため、モデルを制約すること
 - L2正則化（係数ベクトルwのL2ノルム（ユークリッド距離）にペナルティ付加）
 - リッジの制約は強い（＝過学習の危険は少ない）
 - モデルの簡潔さ（0に近い係数の数）と、訓練セットの性能はトレードオフ　→　パラメータalphaで調整
 - alpha ：　デフォルト1.0。最適値はデータセットによる。
 - alpha大 → 制約大（係数wはより0に近づく）（訓練スコア小）
 - alpha小 → 制約小（通常最小二乗法に近づく）
 - 実用上は、まずこのリッジを試すと良い。

```python
# リッジ回帰　x 複雑なデータセット（拡張bostonデータセット（特徴量数104））
from sklearn.linear_model import Ridge
m = Ridge().fit(ax,ay)
# 訓練スコアとテストスコア（デフォルト）
print(m.score(ax,ay))
print(m.score(ex,ey))
```

```python
# -> 訓練スコアとテストスコアの差が縮まった（線形回帰よりは過学習を抑制できている）
# 単純なモデルは訓練スコア低く、テストスコア高い。
```

```python
# リッジ回帰（alpha=10）
m10 = Ridge(alpha=10).fit(ax,ay)
# 訓練スコアとテストスコア（alpha=10）
print(m10.score(ax,ay))
print(m10.score(ex,ey))
```

```python
# リッジ回帰（alpha=10）
m01 = Ridge(alpha=0.1).fit(ax,ay)
# 訓練スコアとテストスコア（alpha=0.1）
print(m01.score(ax,ay))
print(m01.score(ex,ey))
```

```python
# -> alpha=0.1が最もテストスコア高い（スコア向上の詳細は5章で）
```

```python
# 正則化を理解1（alphaを変化させて係数の大きさを見る）
plt.plot(m10.coef_, 'o', label="alpha=10")
plt.plot(m.coef_, 'o', label="alpha=1")
plt.plot(m01.coef_, 'o', label="alpha=0.1")
plt.plot(mL.coef_, 'x', label="Linear")

plt.legend()
plt.ylim(-25,25);
```

```python
# -> 制約が強い（alpha大）ほど、係数（y軸）が0に近づく（線形回帰やalpha0.1では、0から離れている）
```

学習曲線：　モデルの性能をデータサイズの関数として表したもの。縦軸にスコア、横軸にデータサイズ。

```python
# 正則化を理解2（alphaは固定、訓練データ量を変化させる）（学習曲線のサンプル）
mglearn.plots.plot_ridge_n_samples()
```

```python
# -> 訓練スコアはテストスコアより高い（当たり前）
#    リッジ（青）は線形回帰（黄）より訓練スコアが低い（正則化しているため）
#    リッジ（青）は線形回帰（黄）よりテストスコアが高い（訓練データが少ないほど顕著）（線形回帰は400以下で無力）
#    訓練データが十分にあれば、線形回帰もリッジも同じ性能（正則化は重要ではない）
#    訓練データが増えるほど、線形回帰の訓練スコアは下がる（過剰適合が解消される）
```

2.3.3.4 Lasso
 - 線形回帰の一種
 - L1正則化（係数ベクトルwのL1ノルム（絶対値の和）にペナルティ付加）
 - 疎な係数（いくつかが完全に0。（＝いくつかの特徴量を無視）（＝自動的に特徴量を選択））
  - モデルを解釈しやすい（使用した（ゼロでない）特徴量がわかるので、どの特徴量が重要なのかわかる）
 - 正則化パラメータalphaでモデルの複雑さ（正則化の強さ＝係数を0に近づかせる強さ）を調整可能（デフォルト1.0)
 - 未学習を解決するには正則化を弱める（alpha減、max_iter(最大繰り返し数）をデフォルトより増（でないと警告発生））
 - alpha小 → 制約小（通常最小二乗法に近づく）
 - 特徴量数が多く、重要なものは少ないと予想できる場合はLassoを試すと良い。（解釈しやすいモデルが欲しい場合も）
 
ElasticNet：　
 - Lidge + Lasso
 - 両方の正則化パラメータを調整する計算コストがかかる
 - 実用上は最良の結果

```python
# Lasso回帰 x 拡張boston
from sklearn.linear_model import Lasso
ma = Lasso().fit(ax,ay)
# 訓練スコア、テストスコア、使用した（ゼロでない）特徴量の数
print(ma.score(ax,ay)) 
print(ma.score(ex,ey))
print(np.sum(ma.coef_ != 0))
```

```python
# -> 訓練スコアもテストスコアも悪い　=　未学習
#    使用した特徴量の数も104の内、4つしかない
```

```python
# 未学習の解決（alpha減、max_iter増）
ma001 = Lasso(alpha=0.01, max_iter=100000).fit(ax,ay)
# 訓練スコア、テストスコア、使用した（ゼロでない）特徴量の数
print(ma001.score(ax,ay)) 
print(ma001.score(ex,ey))
print(np.sum(ma001.coef_ != 0))
```

```python
# -> 各スコア上昇。使用した特徴量数も4->33に増えた。
#    テストスコアはリッジと同等だが、Lassoは使用した特徴量がわかるので、モデルを理解しやすい
```

```python
# 未学習の解決（alpha減、max_iter増）
ma00001 = Lasso(alpha=0.0001, max_iter=100000).fit(ax,ay)
# 訓練スコア、テストスコア、使用した（ゼロでない）特徴量の数
print(ma00001.score(ax,ay)) 
print(ma00001.score(ex,ey))
print(np.sum(ma00001.coef_ != 0))
```

```python
# -> 過学習発生。使用した特徴量数は104中94（学習しすぎ）。alphaを小さくし過ぎてもダメ。
```

```python
# 正則化を理解3（alphaを変化させて係数の大きさを見る）
plt.plot(ma.coef_, 'o', label="alpha=1")
plt.plot(ma001.coef_, 'o', label="alpha=0.01")
plt.plot(ma00001.coef_, 'o', label="alpha=0.0001")
plt.plot(mL.coef_, 'x', label="Linear")

plt.legend()
plt.ylim(-25,25);
```

```python
# -> 制約が強い（alpha大）ほど、係数（y軸）が0に近づく（線形回帰やalpha0.0001では、0から離れている）
#    いくつかが完全にゼロ（Lassoのため）
```

2.3.3.5 クラス分類のための線形モデル（その他）
- 2クラス分類の予測式：　y = wx + b > 0　（ただし wx = w[0]x[0] + w[1]x[1] + ... )（予測されたyが正： +1クラス、　負：　-1クラス）
- 線形回帰：　yは特徴量xの線形関数
- 線形分類：　決定境界が入力の線形関数（線形の2クラス分類器：　直線や平面や超平面で2クラスを分割）

- 線形モデルを学習するアルゴリズムの違い：
 - 係数と切片の特定の組み合わせと訓練データの適合度を測る尺度（損失関数)（w,bの調整は技術的にできない）
 - 正規化を行うか、またその方法
 
- 線形分類アルゴリズム：　
 - 例：
  - ロジスティック回帰
  - 線形SVM
 - デフォルトでL2正則化あり
 - 正則化パラメータ：　C　（C大→正則化弱（＝モデル複雑化。個々のデータも細かく分類しようとして過学習より）（デフォルト1）
 

```python
# ロジスティック回帰、線形SVM x forgeデータセット
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
x,y = mglearn.datasets.make_forge()
f,a1 = plt.subplots(1,2,figsize=(10,3))
for m,a in zip([LinearSVC(), LogisticRegression()], a1):
    m = m.fit(x,y)
    # 境界プロット
    mglearn.plots.plot_2d_separator(m,x,fill = False, eps=0.5, ax=a, alpha=0.7)
    # 散布図
    mglearn.discrete_scatter(x[:,0], x[:,1], y, ax=a)
    a.set_title(m.__class__.__name__) # モデル名を取得
a1[0].legend();
```

```python
# 線形分類 x Cの影響
mglearn.plots.plot_linear_svc_regularization()
```

```python
# -> C大→　正則化弱（＝モデル複雑化。個々のデータも細かく分類しようとして過学習より）
```

```python
# ロジスティック回帰 x cancerデータセット
from sklearn.datasets import load_breast_cancer
d = load_breast_cancer()
ax,ex,ay,ey = train_test_split(d.data, d.target, stratify=d.target, random_state=42)
mo = LogisticRegression().fit(ax, ay)
# 訓練スコアとテストスコア
print(mo.score(ax,ay))
print(mo.score(ex,ey))
```

```python
# -> 各スコア非常に良い。ただ訓練スコア≒テストスコアのため、未学習より。
```

```python
# ロジスティック回帰 x cancerデータセット （C大→　正則化を弱くする）
mo100 = LogisticRegression(C=100).fit(ax,ay)
# 訓練スコアとテストスコア
print(mo100.score(ax,ay))
print(mo100.score(ex,ey))
```

```python
# -> 各スコアさらに高くなった。
```

```python
# ロジスティック回帰 x cancerデータセット （C小（正則化強化）も試す）
mo001 = LogisticRegression(C=0.01).fit(ax,ay)
# 訓練スコアとテストスコア
print(mo001.score(ax,ay))
print(mo001.score(ex,ey))
```

```python
# -> 各スコア悪化（想定通り）
```

```python
# 正則化を理解4（Cを変化させて係数の大きさを見る）
plt.plot(mo100.coef_.T, 'o', label="C=100")
plt.plot(mo.coef_.T, 'o', label="C=1")
plt.plot(mo001.coef_.T, 'o', label="C=0.01")
plt.legend();

plt.xticks(range(d.data.shape[1]), d.feature_names, rotation=90);
```

```python
# → C大　→　係数はゼロから離れる（正則化弱）
```

```python
# L1正則化を試す（どの特徴量が分類の役に立っているのか知りたい）
for C in [0.001, 1, 100]:
    mo = LogisticRegression(C=C, penalty="l1").fit(ax, ay)
    print("C=",C, "aa=",round(mo.score(ax,ay),2))
    print("C=",C, "ea=",round(mo.score(ex,ey),2))
    plt.plot(mo.coef_.T, 'o', label=C);
plt.xticks(range(d.data.shape[1]), d.feature_names, rotation=90);
plt.legend();
```

```python
# -> 正則化パラメータCによって、正則化の強さと使用する特徴量が決まる
```

2.3.3.6 線形モデルによる多クラス分類
- 線形分類アルゴリズムの多くは2クラス分類のみ可能で、多クラス分類は不可能
- 1対その他アプローチ：
 - 2クラス分類アルゴリズムを多クラス分類アルゴリズムに拡張する一般的な方法
 - 学習：　各クラスに対し、そのクラスと他の全てのクラスを分類する2クラス分類モデルを学習（クラスごとに係数ベクトルwと切片bがある）
 - 予測：　全ての2クラス分類器を実行し、確信度が最高となるクラス分類器が対応するクラスを採用
 - 確信度：　y = w[0]x[0] + w[1]x[1] + ... + b 
- 多クラスロジスティック回帰：　
 - 用いる数学は異なるが、各クラスで係数ベクトルwと切片bがあるのは同じ。同じ方法で予測可能。

```python
# 1対その他アプローチ x 3クラス分類（make_blobsデータセット）
from sklearn.datasets import make_blobs
x,y = make_blobs(random_state=42)
# 散布図
mglearn.discrete_scatter(x[:,0], x[:,1], y);
```

```python
# 線形SVM
m = LinearSVC().fit(x,y)
# 傾きと切片の形状
print(m.coef_.shape)
print(m.intercept_.shape)
```

```python
# -> m.coef_（傾きベクトルw）の各行は各クラス(3)を表し、各列は特徴量数(2)を表す
#    m.intercept_ (切片b)は各クラス(3)に対する切片
```

```python
# 1対その他アプローチ x 3クラス分類（make_blobsデータセット）
# 散布図
mglearn.discrete_scatter(x[:,0], x[:,1], y)
l = np.linspace(-15,15)
for c, i in zip(m.coef_, m.intercept_):
    # 境界描画
    plt.plot(l, -(l * c[0] + i) / c[1])
```

```python
# 真ん中の三角形（どのクラスにも属さない領域）では、最も近い境界線の属するクラスとする（クラス分類式の値が最高となるクラス）
```

```python
# 1対その他アプローチ x 3クラス分類（make_blobsデータセット）
# エリア着色
mglearn.plots.plot_2d_classification(m,x,fill=True,alpha=0.7)
# 散布図
mglearn.discrete_scatter(x[:,0], x[:,1], y)
l = np.linspace(-15,15)
for c, i in zip(m.coef_, m.intercept_):
    # 境界描画
    plt.plot(l, -(l * c[0] + i) / c[1])
```

2.3.3.7 利点、欠点、パラメータ

線形モデルの主要なパラメータ：
- 回帰：　alpha （alpha大　→　モデル単純化）
- 分類：　C　　 （C小　　　→　モデル単純化）

パラメータ調整：
- 特に回帰で重要
- 対数スケールで値を変更するのが一般的
- 正則化をL1にするかL2にするか決める
 - L1：　一部の特徴量だけが重要と分かっている場合。どの特徴量がそのモデルにとって重要か知りたい場合。
 - L2：　デフォルトでおすすめ。

線形モデルの長所・短所：
- 高速（学習も予測も）
- 疎なデータにも対応
- データセットが非常に大きくても対応。サンプル点が100万点→　
 1. ロジスティック回帰 or リッジはsolver='sag'オプションで高速になる可能性

 - 低次元では他のモデルの方が高性能のこともある

2.3.5 決定木
- 分類も回帰も可能

- 純粋：　クラスを1つしか含まない決定木の葉

決定木の構築：
- アルゴリズムは全ての可能なテストの中から、目的変数に対して最も情報量の多いものを選ぶ。

過学習の防止策：
- 事前枝刈り：　木の高さ制限、葉の最大値制限、分割時にそこに含まれる点の最小数を設定
- 事後枝刈り：　普通の枝刈り（sklearnには無い）

決定木クラス：
 - 回帰：　DecisionTreeRegressor
 - 分類：　DecisionTreeClassifier

2.3.5.5 決定木の長所、短所、パラメータ

決定木のパラメータ：
- 事前枝刈りパラメータのみ（max_depth, max_leaf_nodes, min_samples_leaf）

長所：
- 非常に高速
- 可視化可能＋説明容易（結果のモデルを）
- 正規化、標準化が不要（個々の特徴量は独立に処理され、データ分割がデータスケールによらない）
　（＝特徴量ごとにスケールが大きく異なったり、2値特徴量と連続値特徴量が混じってもOK）

- 過学習しやすい（事前枝刈りしても）
- 汎化性能低め（このため、決定木単体でなくアンサンブル（ランダムフォレスト）で用いる）

2.3.6 決定木のアンサンブル

アンサンブル：　
- 複数モデルを組み合わせて強力なモデルを構築する手法
- 分類や回帰に有効とわかっているのは 1. ランダムフォレスト、2. 勾配ブースティング決定木

2.3.6.1 ランダムフォレスト
- 決定木のアンサンブル
- 個々の決定木が異なる方向に過剰適合しても、それらの平均を取れば過剰適合の度合いを減らせる（数学で証明可能）
- 個々の決定木が互いに異なっていなければならないため、構築過程で乱数を導入している
　（1. データの選択、2. 分岐テストに用いる特徴量の選択）
- 本質的に、決定木の利点を残したまま欠点の一部を補っている
- パラメータは全て事前枝刈り用。デフォで十分なことが多いが、調整すると性能向上、大幅時間短縮の場合もある。

ランダムフォレストの構築
- 決定木の数を決定（n_estimators）（実際は数百から数千も）
- 個々の木は互いに独立して構築される（異なる木になるように乱数使用）
- 手順：
 1. データからブートストラップサンプリング（復元抽出＝データ点から重複ありでランダムにデータ点数だけ抽出）
   - データ点数は同じで、一部は重複し、元の1/3ぐらいの値は欠けるイメージ（a,b,c,d -> b,d,d,aとかd,a,c,cとか）
   - 個々の決定木が異なるデータに対して構築される
 2. 得られたデータセットで決定木構築
   - 個々のノードで最適なテスト（質問）を選ぶのでなく、特徴量のサブセットをランダムに選び、
       その中で最適なテストを選択
   - サブセットの大きさ：　max_features（重要）
   - サブセットの選択：　個々のノードで独立に繰り返し行われる
   - 個々の決定木が異なる特徴量に対して分類を行う
   
- max_featuresの設定：
 - n_featuresに設定　→　全ての特徴量を使用（特徴量選択に関する乱数性なし）
 - 1に設定　→　ランダムに選ばれた1つの特徴量だけについて分岐する閾値を探す
 - 値が大　→　過学習より（使用する特徴量が多いため学習は容易。最も識別性の高い特徴量を使用でき、各木が似る）
 - 値が小　→　未学習より（使用する特徴量が少ないため、深くしないと学習が難しめになる）
 
ランダムフォレストの予測
 1. 全ての木に対して予測（回帰は連続値を、分類は各クラスに対する確率を算出）
 2. 平均を採用（回帰は連続値の平均、分類は確率の平均値が最高となるクラスを採用）

長所：
- 非常に強力（分類でも回帰でも、現在最も使われている）
- 頑健
- パラメタ調整しなくても強力
- データスケール変換不要（大規模データもOK)
- 並列処理可能（使用するCPUコア数をn_jobs属性で設定すれば、線形に高速化（2コア使うと時間半分）（-1指定でmax化））

短所：
- モデルの理解が難しい（単一の決定木は可視化できたが）
- ランダムフォレスト内の決定木は構造が深い（特徴量のサブセットを使うため、未学習にならないように深くなる）
- 乱数シードの設定でモデルが大きく変わる可能性がある（本質的にランダムな手法）（決定木数増　→　乱数性減）
- 高次元で疎なデータは苦手（テキストデータなど）（→　線形モデルの方が良い）
- メモリ使用大（線形モデルより）
- 遅い（線形モデルより）

パラメータ：

2.3.7 カーネル法を用いたサポートベクタマシン
- 普通のSVM（線形でない）
- 複雑なモデルに対応（線形SVMを拡張）（超平面のような簡単なモデルではない）
- 分類も回帰も可能（分類：SVC、回帰：SVR）
- 数学的には難解（書籍「統計的学習の基礎」）（実用上は重要ではない）
- 学習：　個々のデータが2クラスの決定境界を表現するのにどの程度重要かを学習
- サポートベクタ：　2クラスの境界周辺の一部のデータで、決定境界を決定するもの。
- 予測：　サポートベクタとデータポイントとの距離と、学習された個々のサポートベクタの重要性で決まる　
- サポートベクタの重要性： SVCのdual_coef_属性
- データポイント間の距離：　ガウシアンカーネルkで測定：　k = exp(-γ||x1-x2||^2)　
 -（ただしx1,x2：データ点。γ:ガウシアンカーネルの幅を制御するパラメータ。|| ... ||：ユークリッド距離 ）

パラメタ調整：
- カーネルの選択：　RBFなど
- C：　個々のデータポイントの重要度（dual_coef_）を制限（線形モデルと同様。正則化用）（デフォルトC=1)
   (C小：　モデル単純、境界大雑把）
- カーネル固有のパラメータ（RBFならgammaのみ）
 - gamma：　ガウシアンカーネルの幅の逆数。点の近さを表すスケールを決定（デフォルトgamma = 1 / n_features)
   （gamma小：　モデル単純、境界大雑把、ガウシアンカーネル直径大（多くの点を近いと判断）、モデル単純、境界大雑把)

長所：
- 様々なデータに使えて強力
- 同質の特徴量向け
- 中規模なデータセット向け
- 特徴量が多くても少なくても（高次元でも低次元でも）複雑な決定境界を生成

短所：
- サンプル数大は苦手（多くて1万まで。10万ではメモリや実行時間的に困難）
- 前処理必要（MinMaxScalerなど）（データのスケールに敏感なため。全ての特徴量の変位が同じスケールであることを要求）
 - 勾配ブースティングなど、ほとんど前処理不要な木系モデルの方がラクではある
- パラメータ設定に敏感
- 検証困難（予測結果の理由を理解、説明するのが困難）

2.3.7.5 SVMのためのデータの前処理
- 全ての特徴量が0~1になるようにスケール変換する手法が多い（MinMaxScaler）

```python
# SVM前処理　全ての特徴量が0~1になるようにスケール変換（MinMaxScalerと同じことを手でやる）
# 訓練データの特徴量ごとに最小値を取得
amin = ax.min(axis=0)
# 訓練データの特徴量ごとにレンジを計算
arange = (ax - amin).max(axis=0)
# 最小値を引いてレンジで割る　＝　最小0, 最大1にスケール変換
axs = (ax - amin) / arange
# 特徴量ごとの最小値（0）と最大値（1）を確認
print(axs.min(axis=0))
print(axs.max(axis=0))
```

```python
# テストセットでも同様（訓練セットの最小値とレンジを用いる）
exs = (ex - amin) / arange
m = SVC()
m.fit(axs, ay) # スケール変換後データ使用
# 訓練スコア・テストスコア
print(m.score(axs, ay)) # スケール変換後データ使用
print(m.score(exs, ey)) # スケール変換後データ使用
```

```python
# -> 訓練スコアとテストスコアが近い　＆　100%から遠い（？）　→　未学習　→　モデルを複雑化する（C大、γ大）
```

```python
# モデルを複雑化（C大）
m = SVC(C=1000)
m.fit(axs, ay)
# 訓練スコア・テストスコア
print(m.score(axs, ay)) # スケール変換後データ使用
print(m.score(exs, ey)) # スケール変換後データ使用
```

```python
# -> モデル複雑化（C大）で精度向上
```

2.3.8 ニューラルネットワーク（ディープラーニング）
- 特定のアプリケーションに向けて注意深く作成されたものが殆ど
- 多層パーセプトロン（MultiLayer Perceptron：MLP）など

多層パーセプトロン(MLP)（＝フィードフォワードNN、またNN）：
- 比較的単純
- 線形モデルを一般化し、決定までに複数のステージで計算するもの
- 線形モデルと同じ予測式：　y = wx + b （ただし wx = w[0] x[0] + w[1] x[1] + ... )
 - x：　入力特徴量
 - w：　学習されたパラメータ（傾き。重み）
 - b：　学習されたパラメータ（切片）
 - y：　出力ラベル（予測結果）
- 線形モデルより強力にするための仕掛け：　個々のユニットの重み付き和の結果に非線形関数（活性化関数）を適用
- 活性化関数：
 - ReLU：　正規化線形関数（y=xのゼロ以下をゼロとする）（デフォルト）
 　（隠れ層1層のrelu：　決定曲線は10の線分から構成される。）
 - tanh：　双曲線正接関数（-1から1に値を押し込める）

パラメータ：
- 隠れ層数：　（最重要）1か2で始めて拡張すればいい。（デフォルト1層）
- 隠れ層のユニット（ノード）数：　（最重要）hidden_layer_sizes（=[10 10]など）入力層と同じが多い。数千より大きくなることはあまりない。（デフォルト100）
- solver：　lbfgsなど
- alpha：　l2ペナルティの強さ（重みを0に近づけてモデルの複雑さを制御）（alpha大→正則化強い）（デフォルト：0.0001）
- 学習繰り返し数：　max_iter（デフォルト200回）

モデルの複雑さ：
- 重み or 係数の数が1つの尺度（隠れノード数が100と1000とでは重みの数が50倍違う（2万100と110万100）

パラメタ調整の一般的な手法：
 1. 過学習できるように大きいネットワークを作成し、タスクがそれを用いて学習可能であることを確認
 2. ネットワークを小さくする or alpha増（正則化強化）　→　汎化性能向上

学習アルゴリズム（パラメータ学習に使用）（どのようにモデルを学習するかの仕組み）：
- adam（デフォルト）：　殆どの場合うまくいく。データスケールに敏感。
- lbfgs：　頑健。モデル大、データ大では遅い。（L-BFGS法：Limited memory BFGS）
 - BFGS：　ブロイデン・フレッチャー・ゴールドファーブ・シャンノ法 (英: Broyden–Fletcher–Goldfarb–Shanno algorithm)
- sgd：　多くのパラメタ調整が必要。

決定境界：

```python
# MLP x cancerデータセット（データのスケーリング（StandardScalerと同じことを手で））
# 訓練セットの特徴量ごとの平均値を算出
am = ax.mean(axis=0)
# 訓練セットの特徴量ごとの標準偏差を算出
ast = ax.std(axis=0)
# 平均を引いて標準偏差で割る（平均0、分散1になる）
axs = (ax - am) / ast
exs = (ex - am) / ast
m = MLPClassifier(random_state=0)
m.fit(axs, ay) # スケーリング後データ使用
# 訓練スコア・テストスコア
print(m.score(axs, ay)) # スケール変換後データ使用
print(m.score(exs, ey)) # スケール変換後データ使用
```

2.4 クラス分類器の不確実性推定
- sklearn関数（sklearn分類器はこの内片方か両方をカバー）：　
 - decision_function：　
   - あるテスト点が陽性である度合い。正：陽性(クラス1)、負：陰性(クラス0)。
   - 形状：　2値分類だけ特殊で、n_samples（＝サンプル数）の一次元のみ。　3値分類以上：(n_samples, n_classes）
   - スケールは様々（確率ではない）（モデルパラメタやデータによる）。
   - デフォルトで使用可能。
 - predict_proba：　
   - 各クラスに属す確率。理解容易。
   - 形状：　(n_samples, n_classes）。（2クラス分類でも(n_samples, 2）で考え方は共通）
   - スケールは0~1（確率なので）で全クラス（全列）足すと1。
   - オプション function='predict_proba' 指定で使用可能。

- モデルが報告する確信度が、実際の正答率と一致している状態を較正されている（calibrated）という。
 - 例：　確信度70%の点に対し、70%の確率で正答できる場合、較正されている。
 - 確信度が実際の正答率と合うかどうかは、モデルやパラメタに依る
  （過学習モデルは間違っていても高い確信度出す。単純なモデルは低い確信度出す）

- ある点の予測がどれくらい確かなのか
- 現実世界では間違えると重大な問題（6章）
 - 偽陽性：　余計な検査を受けるだけ
 - 偽陰性：　深刻な病気が治療されない危険

- sklearn Webサイト(参考）：
 - http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html

```python
# 不確実性推定 x 勾配ブースティング x 合成2次元データセット（circles） 
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_circles
x,y = make_circles(noise=0.25, factor=0.5, random_state=1)
# yが0なら"blue", 1なら"red"としてyを置き換えた配列作成（yは0,1の配列。これをインデックスとする）
yn = np.array(["blue","red"])[y] 
# 分割（ynを入れても整合するようにできている）
ax,ex,ayn,eyn,ay,ey = train_test_split(x,yn,y, random_state=0)                   
m = GradientBoostingClassifier(random_state=0)
m.fit(ax,ayn)
# テストデータ形状
print(ex.shape)
# 決定関数（不確実性推定）
p = m.decision_function(ex)
print(p.shape)
```
